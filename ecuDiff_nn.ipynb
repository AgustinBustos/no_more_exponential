```python
import torch
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
```

```python
x=np.linspace(-2,2)
y=np.exp(x)
px.line(x=x,y=y).show()
```

```python
xt=Variable(torch.linspace(-2,2,50), requires_grad=True)

def mse(t1,t2):
  diff = t1 - t2
  return torch.sum(diff * diff) / diff.numel()
  
sig= torch.nn.Sigmoid()
relu = torch.nn.ReLU()
```

```python
measure=100
w=torch.randn([1,measure],requires_grad=True)
b=torch.randn(measure,requires_grad=True)

w1=torch.randn([measure,1],requires_grad=True)
b1=torch.randn(1,requires_grad=True)

```

```python
def f_hat(x):
  layer1=sig(torch.reshape(x,(-1,1)) @ w+b)  #relu o sig  #b
  result=layer1 @ w1 + b1  #b1
  return result.reshape(-1)


fig1=px.line(x=xt.reshape(-1).detach().numpy(),y=f_hat(xt).detach().numpy())
fig2=px.line(x=x,y=y)
fig3 = go.Figure(data=fig1.data + fig2.data)
fig3.show()
```

```python
def f_hat_prime(x):
  f_hat(x).backward(torch.ones_like(f_hat(x)))
  x.grad.zero_()

  f_hat(x).backward(torch.ones_like(f_hat(x)))
  return x.grad

f_hat_prime(xt)
```

```python

px.line(x=list(xt.reshape(-1).detach().numpy())+list(xt.reshape(-1).detach().numpy()),
        y=list(f_hat(xt).detach().numpy())+list(f_hat_prime(xt).detach().numpy())).show()
```

```python
fig1=px.scatter(x=list(f_hat_prime(xt).detach().numpy()),y=list(f_hat(xt).detach().numpy()))
fig2=px.line(x=np.linspace(0,7),y=np.linspace(0,7))

fig3 = go.Figure(data=fig1.data + fig2.data)
fig3.show()
```

```python
# measure=100
# w=torch.randn([1,measure],requires_grad=True)
# b=torch.randn(measure,requires_grad=True)

# w1=torch.randn([measure,1],requires_grad=True)
# b1=torch.randn(1,requires_grad=True)


optimizer = torch.optim.Adam([w,b,w1,b1], lr=0.01)  


for epoch in range(1000):   
    loss = mse(f_hat(xt), f_hat_prime(xt))
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
fig1=px.scatter(x=list(f_hat_prime(xt).detach().numpy()),y=list(f_hat(xt).detach().numpy()))
fig2=px.line(x=np.linspace(0,7),y=np.linspace(0,7))

fig3 = go.Figure(data=fig1.data + fig2.data)
fig3.show()
fig1=px.line(x=xt.reshape(-1).detach().numpy(),y=f_hat(xt).detach().numpy())
fig2=px.line(x=x,y=y)
fig3 = go.Figure(data=fig1.data + fig2.data)
fig3.show()

px.line(x=list(xt.reshape(-1).detach().numpy())+list(xt.reshape(-1).detach().numpy()),
        y=list(f_hat(xt).detach().numpy())+list(f_hat_prime(xt).detach().numpy())).show()
```
